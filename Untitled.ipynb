```python
def get_unbounded_sets(total, n):
    if n > total:
        return []
    
    if n == total:
        return [[1] * n]
    
    out = []
    for i in range(1, total+1):
        for s in get_unbounded_sets(total-i, n-1):
            out.append([i] + s)
    return out
```
